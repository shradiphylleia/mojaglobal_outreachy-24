Importing required libraries

```python
import pandas as pd
import seaborn as sns
import geopandas as gpd
import matplotlib.pyplot as plt
```

Reading the geojson data

```python
def rdgeojson(link):
    try:
        return gpd.read_file(link)
    except Exception as e:
        print("An error was encoutered:",e)
```

```python
df_cntry=rdgeojson("/workspaces/mojaglobal_outreachy-24/IND_AL2_India.json")
df_state=rdgeojson("/workspaces/mojaglobal_outreachy-24/IND_AL4_Himachal Pradesh_GEZ.json")
```

```python
df_state
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
df_cntry.plot(ax=ax,edgecolor="black")
plt.show()
```
